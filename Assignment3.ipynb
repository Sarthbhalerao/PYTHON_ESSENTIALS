 **Q1. Write down a program in Python for Opening a File and Writing " I Love LetsUpgrade" And close
it, and read it back again, and then append some data to it and close it.**

**Answer:-**

```
 file = open('mbox.txt','w') # open a file in write mode
file.write('I love LetsUpgrade.\n') # Wriiten in file
file.close()
file = open('mbox.txt','r')
file = open('mbox.txt','a') # Appending data into file.
file.write('Hello all\n')
file.close()
file = open('mbox.txt','r')
print(file.read())
```

 **Q2. Write a function which can return a Factorial of any numbers as INT, given in the argument.**

**Answer-**

```
 def factorial(num):
    if num == 1:
       return num
    elif num < 1:
       return ("NA")
    else:
       return num*factorial(num-1)
factorial(int(input('Enter the number:')))
```
