 **Q1. Try 5  different functions of the string in python**

**Answer:-**

```
 
#creating strings
 
str1 = "Hello, Welcome to Python.."
str2 = "Here is my first Assignment" 
print(str1) 
print(str2)
```

1)   index()   =>  It Searches the string for a specified value and returns the position of where it was found.
 
Syntax-
string.index(value, start, end)

```
 
print(str1) 
X = str1.index("P") 
print(X) 
print(str2) 
Y = str2.index("s", 2,10)
print(Y)
```

2) capitalize  => Converts the first character to upper case
 
Syntax-
string.capitalize()

```
print(str1)
print(str2) 
a = str1.capitalize() 
print(a) 
b = str2.capitalize() 
print(b)
```

3) split() => Splits the string at the specified separator, and returns a list.
Syntax-
string.split()

```
print(str1) 
c = str1.split()
print(c)
print(str2) 
d = str2.split()
print(d)
```

4) upper() => Converts a string into upper case
 
syntax- string.upper()

```
print(str1)
f= str1.upper()
print(f)
 
print(str2)
g= str2.upper()
print(g)
```

5) swapcase() => Swaps cases, lower case becomes upper case and vice versa. 
syntax-  string.swapcase()

```
print(str1)
i= str1.swapcase()
print(i)
 
print(str2) 
j= str2.swapcase()
print(j)
```

6) count() => Returns the number of times a specified value occurs in a string
 
syntax- string.count(value, start, end)

```
print(str1)
k=str1.count("o", 1,15)
print(k)
 
print(str2)
L=str2.count("first", 1,25)
print(L)
```

7) isdecimal() =>
Returns True if all characters in the string are decimals
syntax- string.isdecimal()

```
print(str1)
str1.isdecimal()
```

 **Q3. Experiment with at least 5 different functions of dictionary.**

**Answer**

```
#creating a dictionary
dt_1 = { "apple":2,"mango":5,"strawberry":3,"banana":9} 
dt_1
```

1) clear() => Removes all the elements from the dictionary. 
 
synrax - dictionary.clear()

```
dt_1.clear()
print(dt_1)
```

2) keys() => Returns a list containing the dictionary's keys. 
 
syntax - dictionary.keys()

```
 
dt = { "apple":2,"mango":5,"strawberry":3,"banana":9, 17:8} 
dt
```

```
z=dt.keys()
print(z)
```

 3) values() => Returns a list of all the values in the dictionary. 
 
syntax - dictionary.values()

```
y=dt.values()
print(y)
```

 4) pop() => Removes the element with the specified key.
 
syntax - dictionary.pop(keyname, defaultvalue)

```
dtt= { "apple":2,"mango":5,"strawberry":3,"banana":9, 17:8} 
dtt
```

```
dtt.pop("mango")
dtt
```

5) update() => Updates the dictionary with the specified key-value pairs. 
 
syntax - dictionary.update(iterable)

```
 car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
 
car.update({"color": "White"})
 
print(car)
```
