**Q1. Use IF ELSE and ELIF to write a program in python for your Report Card**

**Answer:-**

```
#Grading system using IF ELSE and ELIF 
 
score = float(input("Enter Your Score between 1 to 10:"))
if score > 10:
    print("wrong score") 
elif score >9 and score <=10:
    print("Congratulation..!!") 
    print("You are passed with grade A") 
elif score >8 and score <9:
    print("Congratulation..!!") 
    print("You are passed with grade B") 
elif score >7 and score <8:
    print("Congratulation..!!") 
    print("You are passed with grade C")
elif score >6 and score<=7:
    print("Congratulation..!!") 
    print("You are passed with grade D")
else:
    print("Opss.. You failed..") 
    print("Better luck next time.. ")
```

**Q2. Use for loop to print prime numbers between 1 to 1000.**

**Answer:-**

```
# Check prime numbrs from 1 to 1000
for num in range(1,1000 + 1):  
   if num > 1:  
       for i in range(2,num):  
           if (num % i) == 0:  
               break  
       else:  
           print(num)
```

**Q3.Write a program for printing the table from 1 to 10 using Nested for Loop.**

**Answer:-**

```
for i in range(1,11):
    print("Table of", i, "is:")
    print("") 
    for j in range(1,11):
        print(i*j)
        print("")
```
